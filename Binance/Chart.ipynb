```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

folder_path = r'C:\Users\Admin\OneDrive\Desktop\Crypto\Binance\Data'
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
```

```python
for file in csv_files:

    file_path = os.path.join(folder_path, file)

    data = pd.read_csv(file_path)
    coin_name = os.path.splitext(file)[0]

    data.rename(columns={"Unnamed: 0": "openTime"}, inplace=True)
    data['openTime'] = pd.to_datetime(data['openTime'])


    fig, ax1 = plt.subplots(figsize=(10, 5))

    ax1.plot(data['openTime'], data['close'], color='blue', label='Close Price')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Close Price', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
    ax1.xaxis.set_major_locator(mdates.AutoDateLocator())
    ax2 = ax1.twinx()
    ax2.bar(data['openTime'], data['volume'], alpha=0.3, color='orange',label='Volume')
    ax2.set_ylabel('Volume', color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')
    # nh dng trc x ca ax2 tng t ax1
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
    ax2.xaxis.set_major_locator(mdates.AutoDateLocator())
    plt.title(f'{coin_name}')
    fig.tight_layout()
    plt.show()
```
