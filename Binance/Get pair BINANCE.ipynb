```python

import pandas as pd
import datetime
import time
import requests
import numpy as np
```

```python
import requests
ticker = []
def get_all_binance_tickers():
    url = 'https://api.binance.com/api/v3/exchangeInfo'

    response = requests.get(url)
    exchange_info = response.json()
    tickers = [symbol['symbol'] for symbol in exchange_info['symbols']]
    return tickers
# Gọi hàm để lấy danh sách các ticker
all_binance_tickers = get_all_binance_tickers()
# In ra danh sách các ticker
for tick in all_binance_tickers:
    if 'USDT' in tick :
        ticker.append(tick)
print(ticker)
```

```python

def get_binance_data_by_requests(ticker,interval):
      """
      interval: str tick interval - 4h/1h/1d ...
      """
      columns = ['open_time','open', 'high', 'low', 'close', 'volume','close_time', 'qav','num_trades','taker_base_vol','taker_quote_vol', 'ignore']
      usecols=['open', 'high', 'low', 'close', 'volume', 'qav','num_trades','taker_base_vol','taker_quote_vol']

      df = pd.DataFrame()
      print(f'Downloading {interval} {ticker} ohlc-data ...', end=' ')
      while True:
          url = f'https://www.binance.com/api/v3/klines?symbol={ticker}&interval={interval}'
          data = pd.DataFrame(requests.get(url, headers={'Cache-Control': 'no-cache', "Pragma": "no-cache"}).json(), columns=columns, dtype=np.float64)    
          start = int(data.open_time.tolist()[-1])+1
          data.index = [pd.to_datetime(x, unit='ms').strftime('%Y-%m-%d %H:%M:%S') for x in data.open_time]
          data = data[usecols]
          df = pd.concat([df, data], axis=0)
          print('Done.')
          time.sleep(0.5)  # Để tránh vượt quá giới hạn tốc độ API
          return df
```

```python

for providerId in ticker[:]:
    try :
        df2 = get_binance_data_by_requests(ticker=providerId, interval='1d')
        filename = f"{providerId}.csv"
        df2.to_csv(filename)
        print(f"Đã lưu dữ liệu cho {providerId} vào {filename}")
    except Exception as e:
        print(f"Lỗi khi tải dữ liệu cho {providerId}: {str(e)}")


```
